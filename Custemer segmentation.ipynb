# DATA Scaling 

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
df_transform=scalar.fit_transform(df)
```

```python

df_transform
```

```python
## principle component analysis(dimentiality reduction(linear))
## this field need to be changed according to model perfomence

from sklearn.decomposition import PCA
pca=PCA(n_components=2)
princ_components=pca.fit_transform(df_transform)
princ_components
```

```python
## converting the pca data into a df 
df_pca= pd.DataFrame(data=princ_components,columns=['PC1','PC2'])
```

```python
df_pca
```

```python
# plotting the pca's here
plt.figure(figsize=(10,8))
sns.scatterplot(x='PC1',y='PC2', data= df_pca)
```

```python
##important liberaries for next steps
from sklearn.cluster import KMeans,AgglomerativeClustering,DBSCAN,SpectralClustering
from sklearn.mixture import GaussianMixture
import warnings
warnings.filterwarnings("ignore") 
""""""
```

# HyperParameter Tunning

```python
# finding the optimle value of k for clustering
inertia=[]
rangevalue=range(1,15)
for i in rangevalue:
    kmean=KMeans(n_clusters=i)
    kmean.fit_predict(pd.DataFrame(df_transform))
    inertia.append(kmean.inertia_)

plt.plot(rangevalue,inertia,'bx-')
plt.xlabel('Values of K') 
plt.ylabel('Inertia') 
plt.title('The Elbow Method using Inertia') 
plt.show()
```

## Model Building using K-Means


```python
# elbow as k=4
kmeans_Model=KMeans(4)
kmeans_Model.fit_predict(pd.DataFrame(df_transform))
```

```python
## concanate the clusters to pca data
pca_kmeans=pd.concat([df_pca,pd.DataFrame({'cluster':kmeans_Model.labels_})],axis=1)
```

```python
pca_kmeans.head()
```

```python
## Visualizing clusters throgh the pca

plt.figure(figsize=(8,8))
sns.scatterplot(data=pca_kmeans,x='PC1',y='PC2',hue=pca_kmeans['cluster'],palette=['red','green','blue','black'])
```

```python
# find all cluster centers
cluster_centers = pd.DataFrame(data=kmeans_Model.cluster_centers_,columns=[df.columns])
cluster_centers
```

```python
## retriving inverse transforming the data to get the original one
cluster_centers = scalar.inverse_transform(cluster_centers)
cluster_centers = pd.DataFrame(data=cluster_centers,columns=[df.columns])
cluster_centers
```

```python
# Creating a target column "Cluster" for storing the cluster segment
cluster_df = pd.concat([df,pd.DataFrame({'Cluster':kmeans_Model.labels_})],axis=1)
cluster_df
```

# Training and Testing the model accuracy using decision tree

```python
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn import tree
from sklearn import metrics
```

```python
x= cluster_df.drop('Cluster',axis=1)
y=cluster_df['Cluster']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
## Training Decision tree Classifier
model= DecisionTreeClassifier(criterion="entropy")
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

```

```python
#Confusion_Matrix
print(metrics.confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

## Saving the Decision tree model for future prediction

```python
filename = 'final_model.sav'
pickle.dump(model, open(filename, 'wb'))
```

```python
# load the model from disk
loaded_model = pickle.load(open(filename, 'rb'))
result = loaded_model.score(x_test, y_test)
print(result,'% Acuuracy')
```
